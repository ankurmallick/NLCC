```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
plt.style.use('publication')
```

```python
with open('multiproc_times.pkl', 'rb') as f:
    scheme_times = pickle.load(f)
```

```python
schemes_list = ['SR','LT','One','Unc','Rep', 'BCC'] #Exclude BCC for thesis 
```

```python
colors = {'One':'Black','Unc':'Orange','Rep':'Brown', 'BCC': 'Purple', 'LT':'Blue','SR':'Green'}
```

```python
plt.figure()
for scheme in schemes_list:
    time_arr = np.array(scheme_times[scheme])
    time_avg = np.mean(time_arr)*1000
    time_std = np.std(time_arr)*1000
    plt.bar(scheme,time_avg,yerr=time_std,color=colors[scheme],ecolor='Magenta',capsize=10)
plt.ylabel('Time (in milliseconds)')
plt.savefig('Plot0_multiproc.pdf')
plt.show()
```
